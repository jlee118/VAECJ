```python
import pandas as pd

apple = pd.read_csv('apple_stock.csv')
apple.plot(x='t', y='Close', figsize=[8,5])
```

```python
train = apple[apple['t'] < 200]['Close']
test = apple[apple['t'] >= 200]['Close']
```

```python
import numpy as np
window=20
X_train = np.asarray([train[i:i+window] for i in range(len(train)-window - 1)])[:,:,np.newaxis]
Y_train = np.asarray([train[i:i+window] for i in range(1, len(train)-window)])[:,:,np.newaxis]
X_test = np.asarray([test[i:i+window] for i in range(len(test)-window-1)])[:,:,np.newaxis]
Y_test = np.asarray([test[i:i+window] for i in range(1, len(test)-window)])[:,:,np.newaxis]
```

```python
from vrnngan import VRNNGRUGAN
from tensorflow import keras
import os
keras.backend.clear_session()

timesteps = X_train.shape[1]
feature_space = X_train.shape[2]
latent_dim = 64
model = VRNNGRUGAN(feature_space, latent_dim, timesteps)
model.compile(keras.optimizers.Adam(lr=0.0001), keras.optimizers.Adam(lr=0.00001))
model.fit(X_train, X_train, batch_size= 6, epochs=100)
model.save_weights('vrnngan_apple_sample2/vrnngan_apple_sample2')
```

```python
def np_kl_gauss(posterior_means, prior_means, posterior_log_var, prior_log_var):
    kl = prior_log_var - posterior_log_var + (np.exp(posterior_log_var) + 
                                       np.square(posterior_means - prior_means)) / np.exp(prior_log_var) - 1
    kl = 0.5 * np.sum(kl)
    return kl
```

# Detection Algorithm 2

```python
from scipy.stats import norm


def detect_cp_cdf(model, alpha, ground_truth, window_size, offset):
    _, _, _, _, _, _, output_mean, output_log_var = model.predict(X_test)
    mean = output_mean[:,-1,:].flatten()
    std = np.exp(output_log_var[:,-1,:].flatten())
    p_values = 1 - norm.cdf(ground_truth, mean, std)
    tail_filter = lambda x: 1 - x if x > 0.5 else x
    p_values = np.fromiter((map(tail_filter, p_values)), dtype=np.float)
    cpd = np.argwhere(p_values < alpha)
    return cpd + window_size + offset
```

```python
cp = detect_cp_cdf(model, 0.15, Y_test[:,-1,:].flatten(), 20, 200).flatten()
changepoints = apple['Close'][cp]
```
